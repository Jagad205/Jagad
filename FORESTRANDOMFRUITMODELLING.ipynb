```python
import pickle
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = dfalas.drop('name',axis=1)
y = dfalas['name']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
yencode = label_encoder.fit_transform(y)
yencode
```

spilt dataset


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, yencode, random_state=20)
```

```python
x_train
```

```python
y_train.shape
```

training mode

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
hutan = RandomForestClassifier()
```

```python
hutan.fit(x_train,y_train )
```

```python
train_score = hutan.score(x_train, y_train)
testing_score = hutan.score(x_test, y_test)
print(f"Score Training : {train_score:4f}")
print(f"Score Testing : {testing_score:4f}")
```

```python
y_predict1 = hutan.predict( x_test)
y_predict1
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_predict1)
```

```python
np.sqrt(mean_squared_error(y_test, y_predict1))
```
